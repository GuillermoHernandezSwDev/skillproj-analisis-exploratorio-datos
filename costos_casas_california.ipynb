```python
new_df = data_frame.dropna()

new_df.info()
```

```python
#Convertir la carcterstica categrica a numrica
#Proximidad al oceano

new_df["ocean_proximity"]
```

```python
new_df["ocean_proximity"].value_counts()

# 1, 2, 3, 4, 5
```

```python
#Dummies / One-Hot Encoding
# NEARBAY INLAND NEAR OCEAN
#   1       2       0
#   0       0       1

dummies = pd.get_dummies(new_df["ocean_proximity"], dtype=int)
```

```python
new_df_w_dummies = new_df.join(dummies)
```

```python
new_df_w_dummies.head(10)
```

```python
new_df_w_dummies.drop(["ocean_proximity"], axis=1)
```

```python
clean_df = new_df_w_dummies.drop(["ocean_proximity"], axis=1)
```

```python
clean_df.head(10)
```

```python
clean_df.info()
```

```python
#Anlisis, nuevas caractersticas

clean_df.corr()
```
